```python
import pandas as pd

#Load the dataset
df = pd.read_csv('/home/ogen/Desktop/practice/data.csv')

#Display the first few rows
df.head()

#Determine the shape of the dataset
print("Shape of the dataset:", df.shape)

#check for missing values
print("Missing values:\n", df.isnull().sum())

#Display basic statistics for the 'amount' column
print("Statistics for'amount' column:\n", df['amount'].describe())



import matplotlib.pyplot as plt

#Plot a bar graph of transaction types 
transaction_counts = df['type'].value_counts()
plt.bar(transaction_counts.index, transaction_counts.values)
plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.title('Count of Each Transaction Type')
plt.show()
#Plot a histogram of transaction amounts
plt.hist(df['amount'], bins=30)
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Transaction Amounts')
plt.show()

#Calculate the number of fraudulent and non-fraudulent transactions
fraud_counts = df['isFraud'].value_counts()

#Plot a pie chart
plt.pie(fraud_counts, labels=['Non-Fraudulent', 'Fraudulent'], autopct='%1.1f%%')
plt.title('Proportion of Fraudulent vs. Non-Fraudulent Transactions')
plt.axis('equal')
plt.show()

#Group the data by transaction type and isFraud column, and calculate the count 
fraud_by_type = df.groupby(['type', 'isFraud']).size().unstack()

#Plot the bar graph
fraud_by_type.plot(kind='bar', stacked=True)
plt.xlabel('Transaction Type')
plt.ylabel('count')
plt.title('Number of Fraudulent Transactions by Transaction Type')
plt.legend(['Non-Fraudulent', 'Fraudulent'])
plt.show()

#Plot the scatter plot
plt.scatter(df['oldbalanceOrg'], df['newbalanceOrig'])
plt.xlabel('Old Balance')
plt.ylabel('New Balance')
plt.title('Balance Analysis')
plt.show()

#Calculate the number of flagged fraud transactions
flagged_fraud_counts = df['isFlaggedFraud'].value_counts()

#Plot a bar chart
plt.bar(flagged_fraud_counts.index, flagged_fraud_counts.values)
plt.xlabel('Flagged is Fraud')
plt.ylabel('count')
plt.title('Number of Flagged Fraud Transactions')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```
